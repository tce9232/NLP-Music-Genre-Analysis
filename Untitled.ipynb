```python
def scrub_lyrics(df):
    lyricslist = []
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    for index, lyrics in enumerate(df['lyrics']):
        # To lowercase
        lyrics = lyrics.lower()
        # Remove whitespace (including new line characters)
        lyrics = re.sub(r'\s\s+', ' ', lyrics)
        #     Remove characters beyond Basic Multilingual Plane (BMP) of Unicode:
    #     lyrics = ''.join(c for c in lyrics if c <= '\uFFFF')
    #     lyrics = re.sub(r'[^\w\s]', '', lyrics)
        lyrics = re.sub(r'\n', ' ', lyrics)
    #     lyrics = re.sub(r'\d+', 'twitterdigit', lyrics)
        lyrics = re.sub("[^a-zA-Z ]+", "", lyrics)
        lyricslist.append(lyrics)
    df['scrubbedlyrics'] = lyricslist
    return df

lyrics = scrub_lyrics(lyrics)
```

```python
lyrics_count = pd.DataFrame()

def concatenate_list_data(list):
    result= ''
    for element in list:
        result = result + " " + str(element)
    return result

for genre in lyrics['genre'].unique():
    for decade in lyrics['decade'].unique():
        data_set = []
        from collections import Counter 
        for lyrics_data in lyrics[(lyrics['genre'] == genre) & (lyrics['decade'] == decade )]['scrubbedlyrics']:
            data_set.append(lyrics_data)

        data_set = concatenate_list_data(data_set)

        # split() returns list of all the words in the string 
        split_it = data_set.split() 

        # Pass the split_it list to instance of Counter class. 
        Counter = Counter(split_it) 

        # most_common() produces k frequently encountered 
        # input values and their respective counts. 
        most_occur = Counter.most_common(10000) 
        most_occur = pd.DataFrame(most_occur, columns=['word', 'count'])
        most_occur['genre'] = genre
        most_occur['decade'] = decade
        print(most_occur.head())
        lyrics_count = lyrics_count.append(most_occur)
```
